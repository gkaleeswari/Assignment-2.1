```python
import os
import random
import string

# Function to generate a random string of given length
def generate_random_string(length=100):
    letters_and_digits = string.ascii_uppercase + string.digits
    return ''.join(random.choice(letters_and_digits) for i in range(length))

# Sizes of files to create in bytes
file_sizes = {
    '1GB': 1 * 1024 * 1024 * 1024,
    '2GB': 2 * 1024 * 1024 * 1024,
    '3GB': 3 * 1024 * 1024 * 1024,
    '4GB': 4 * 1024 * 1024 * 1024,
    '5GB': 5 * 1024 * 1024 * 1024
}

# Directory path
directory_path = 'large_random_files/'

# Ensure the directory exists
if not os.path.exists(directory_path):
    os.makedirs(directory_path)

# Generate files of specified sizes
for file_name, size in file_sizes.items():
    file_path = os.path.join(directory_path, f'random_strings_{file_name}.txt')
    with open(file_path, 'w') as file:
        file_size = 0
        while file_size < size:
            random_string = generate_random_string() + '\n'
            file.write(random_string)
            file_size += len(random_string)

print("Files have been created successfully.")
```

```python
import os
import concurrent.futures

# Directory path where the files are stored
directory_path = 'large_random_files/'

# File names
file_names = [
    'random_strings_1GB.txt',
    'random_strings_2GB.txt',
    'random_strings_3GB.txt',
    'random_strings_4GB.txt',
    'random_strings_5GB.txt'
]

def convert_file_to_uppercase(file_name):
    file_path = os.path.join(directory_path, file_name)

    # Read the contents of the file
    with open(file_path, 'r') as file:
        content = file.read()

    # Convert the content to uppercase
    content_upper = content.upper()

    # Write the uppercase content back to the file
    with open(file_path, 'w') as file:
        file.write(content_upper)

    print(f"{file_name} has been converted to uppercase successfully.")

# Use ThreadPoolExecutor to convert files in parallel
with concurrent.futures.ThreadPoolExecutor() as executor:
    executor.map(convert_file_to_uppercase, file_names)
```

```python
import os

# Directory path where the files are stored
directory_path = 'large_random_files/'

# File names
file_names = [
    'random_strings_1GB.txt',
    'random_strings_2GB.txt',
    'random_strings_3GB.txt',
    'random_strings_4GB.txt',
    'random_strings_5GB.txt'
]

# Function to convert contents of a file to uppercase
def convert_to_uppercase(file_path):
    temp_file = file_path + '.tmp'  # Temporary file path

    try:
        # Open original file for reading and temporary file for writing
        with open(file_path, 'r', encoding='utf-8') as original_file, \
             open(temp_file, 'w', encoding='utf-8') as temp_file:

            # Process line by line to avoid memory issues
            for line in original_file:
                temp_file.write(line.upper())

        # Rename temporary file to original file path
        os.replace(temp_file.name, file_path)
        print(f"{file_path} converted to uppercase successfully.")

    except FileNotFoundError:
        print(f"File not found: {file_path}")
    except IOError as e:
        print(f"Error occurred while processing {file_path}: {e}")

# Convert contents of each file to uppercase
for file_name in file_names:
    file_path = os.path.join(directory_path, file_name)
    convert_to_uppercase(file_path)

print("All files have been converted to uppercase successfully.")
```

```python
from pytube import YouTube
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from oauth2client.tools import argparser

# Define API key and YouTube API service
API_KEY = 'YOUR_YOUTUBE_API_KEY'
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'

def youtube_search(query, max_results=10):
    youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=API_KEY)

    search_response = youtube.search().list(
        q=query,
        part='id,snippet',
        maxResults=max_results
    ).execute()

    videos = []
    for search_result in search_response.get('items', []):
        if search_result['id']['kind'] == 'youtube#video':
            videos.append(search_result)

    return videos

def download_youtube_videos():
    search_query = "Machine Learning"
    num_videos = 10

    try:
        search_results = youtube_search(search_query, num_videos)

        for result in search_results:
            video_id = result['id']['videoId']
            video = YouTube(f"https://www.youtube.com/watch?v={video_id}")
            stream = video.streams.get_highest_resolution()
            print(f"Downloading: {video.title}...")
            stream.download(output_path="./downloads")
            print(f"{video.title} downloaded successfully!")

    except HttpError as e:
        print(f"An HTTP error {e.resp.status} occurred: {e.content}")
    except Exception as e:
        print(f"Error occurred: {str(e)}")

if __name__ == "__main__":
    download_youtube_videos()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances

# Set random seed for reproducibility
np.random.seed(42)

# Generate random data
data = pd.DataFrame({
    'col1': np.random.uniform(low=-10, high=10, size=500),
    'col2': np.random.uniform(low=-10, high=10, size=500),
    'col3': np.random.uniform(low=-10, high=10, size=500),
    'col4': np.random.uniform(low=-10, high=10, size=500),
    'col5': np.random.uniform(low=10, high=20, size=500),
    'col6': np.random.uniform(low=10, high=20, size=500),
    'col7': np.random.uniform(low=10, high=20, size=500),
    'col8': np.random.uniform(low=10, high=20, size=500),
    'col9': np.random.uniform(low=-100, high=100, size=500),
    'col10': np.random.uniform(low=-100, high=100, size=500)
})

# Initialize list to store SSE (Sum of Squared Errors) for each K
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    sse.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), sse, marker='o', linestyle='--')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Example with optimal K=4 (as determined from the elbow method)
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(data)
labels_kmeans = kmeans.labels_

# Plotting the distance metric graph (using pairwise distances)
distances_kmeans = pairwise_distances(data, metric='euclidean')
plt.figure(figsize=(10, 8))
plt.imshow(distances_kmeans, interpolation='nearest')
plt.colorbar()
plt.title('Pairwise Distance Matrix - K-Means Clustering')
plt.show()

# Calculate the linkage matrix
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(data, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Example with optimal number of clusters (from dendrogram)
n_clusters_hierarchical = 4
hierarchical = AgglomerativeClustering(n_clusters=n_clusters_hierarchical)
labels_hierarchical = hierarchical.fit_predict(data)

# Plotting the distance metric graph (using pairwise distances)
distances_hierarchical = pairwise_distances(data, metric='euclidean')
plt.figure(figsize=(10, 8))
plt.imshow(distances_hierarchical, interpolation='nearest')
plt.colorbar()
plt.title('Pairwise Distance Matrix - Hierarchical Clustering')
plt.show()
```

```python
# Generate random data for Question 15
data_q15 = pd.DataFrame(np.random.uniform(low=5, high=10, size=(500, 5)),
                        columns=[f'col{i}' for i in range(1, 6)])
from scipy.stats import ttest_1samp

alpha = 0.05
for col in data_q15.columns:
    t_stat, p_val = ttest_1samp(data_q15[col], 0)  # Perform t-test against mean of 0
    if p_val < alpha:
        print(f"Column {col}: p-value = {p_val} (Reject null hypothesis)")
    else:
        print(f"Column {col}: p-value = {p_val} (Fail to reject null hypothesis)")
from scipy.stats import wilcoxon

for col in data_q15.columns:
    w_stat, p_val = wilcoxon(data_q15[col] - 5)  # Perform Wilcoxon signed rank test against median of 5
    if p_val < alpha:
        print(f"Column {col}: p-value = {p_val} (Reject null hypothesis)")
    else:
        print(f"Column {col}: p-value = {p_val} (Fail to reject null hypothesis)")
from scipy.stats import ttest_ind, ranksums

# Two Sample t-test
t_stat, p_val_t = ttest_ind(data_q15['col3'], data_q15['col4'])
if p_val_t < alpha:
    print(f"Two Sample t-Test: p-value = {p_val_t} (Reject null hypothesis)")
else:
    print(f"Two Sample t-Test: p-value = {p_val_t} (Fail to reject null hypothesis)")

# Wilcoxon Rank Sum Test (Mann-Whitney U test)
w_stat, p_val_w = ranksums(data_q15['col3'], data_q15['col4'])
if p_val_w < alpha:
    print(f"Wilcoxon Rank Sum Test: p-value = {p_val_w} (Reject null hypothesis)")
else:
    print(f"Wilcoxon Rank Sum Test: p-value = {p_val_w} (Fail to reject null hypothesis)")


```
